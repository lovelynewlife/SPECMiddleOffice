```python
has_energy_index = cdf[cdf['energy_peak_result']>0]['index'].iloc[0]
has_energy_index
```

```python
_, has_energy_location, _ = SPEC.op.group.get_one_result_location(benchmark_name, FileType.HTML, has_energy_index)
with open(has_energy_location, "r") as file:
    has_energy_content = file.read()

HTML(has_energy_content)
```

```python
import bs4
import re
template_soup = bs4.BeautifulSoup(has_energy_content)
```

```python
title_info_table = template_soup.find(name="table", attrs={'class': 'titlebarcontainer'})
title_info_table
```

```python
title_info_table_row1, title_info_table_row2 = title_info_table.find_all(name='tr')
sys_info, base_result_summary = title_info_table_row1.find_all(name="td")
sys_info, base_result_summary = sys_info.find_all(name='p'), base_result_summary.find_all(name='p')
sys_info, base_result_summary
```

```python
trim_empties = re.compile("\s")
```

```python
hardware_vendor, system = (trim_empties.sub(" ", elem.getText()).strip() for elem in sys_info)
hardware_vendor = re.search(r"(.*)\(", hardware_vendor).group(1)
hardware_vendor = hardware_vendor.strip()
hardware_vendor, system
```

```python
system_info = re.search(r"(.*)\((.*)\)", system)
if system_info is not None:
    system_name, system_params = system_info.group(1).strip(), system_info.group(2).strip()
    system_name, system_params
else:
    system_name = system
    system_params = None
system_name, system_params
```

```python
base_results = (trim_empties.sub(" ", elem.getText()).strip() for elem in base_result_summary)
base_result, energy_base_result = (re.findall(r".*=(.*)", elem)[0].strip() for elem in base_results)
base_result, energy_base_result
```
